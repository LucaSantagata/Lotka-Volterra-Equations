```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

figure(figsize=(13, 10), dpi=80)

#risolviamo l'equazione differenziale: dy(t)/dt = -k y(t)


# definisco la funzione che indica come è fatta la mia derivata dy/dt (k è un paramentro)
def model(N,t,alpha):
    dNdt = alpha * N
    return dNdt

# condizione iniziale
N_0 = 100

# vettore temporale(indica quanti step temporali voglio avere)
t = np.linspace(0,20)

# risolvo l'equazione per diversi valori di k
alpha_1 = 1
N_1 = odeint(model,N_0,t,args=(alpha_1,))
alpha_2 = 0
N_2 = odeint(model,N_0,t,args=(alpha_2,))
alpha_3 = -0.1
N_3 = odeint(model,N_0,t,args=(alpha_3,))

# grafico le soluzioni
plt.plot(t,N_1,'r-',linewidth=2,label='alpha_1 = 1')
plt.plot(t,N_2,'b-',linewidth=2,label='alpha_2 = 0')
plt.plot(t,N_3,'g-',linewidth=2,label='alpha_3 = -0.1')

plt.xlabel('time')
plt.ylabel('N(t)')
plt.legend()
plt.xlim(0,20)
plt.ylim(0,700)
plt.show()
```
