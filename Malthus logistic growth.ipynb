```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

#risolviamo l'equazione differenziale: dy(t)/dt = -k y(t)
from matplotlib.pyplot import figure

figure(figsize=(13, 10), dpi=80)


# definisco la funzione che indica come è fatta la mia derivata dy/dt (k è un paramentro)
def model(N,t,alpha, K):
    dNdt = alpha * N * (1- N/K)
    return dNdt

# condizione iniziale
N_0 = 100

# vettore temporale(indica quanti step temporali voglio avere)
t = np.linspace(0,300)

# risolvo l'equazione per diversi valori di k
alpha_1 = 0.1
K_1=500
N_1 = odeint(model,N_0,t,args=(alpha_1,K_1,))

alpha_2 = 0.1
K_2=10
N_2 = odeint(model,N_0,t,args=(alpha_2,K_2))

alpha_3 = 0.1
K_3= 800
N_3 = odeint(model,N_0,t,args=(alpha_3,K_3))

#grafico le soluzioni
plt.plot(t,N_1,'r-',linewidth=2,label= 'K = 500')
plt.plot(t,N_2,'b-',linewidth=2,label='K = 10')
plt.plot(t,N_3,'g-',linewidth=2,label='K = 800')

plt.xlabel('time')
plt.ylabel('N(t)')
plt.legend()
plt.xlim(0,350)
plt.ylim(0,1000)
plt.title("alfa = 0.1")
plt.show()
```
