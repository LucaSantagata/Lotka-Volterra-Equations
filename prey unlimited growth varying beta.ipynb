```python
import numpy as np
import scipy.linalg as la
import math
from scipy import integrate
import matplotlib.pyplot as plt
import pylab as p
import math

from matplotlib.pyplot import figure

figure(figsize=(13, 10), dpi=80)

"""DEFINIAMO I PARAMETRI"""

a = 1     #1. alfa
b = 0.04     #0.1 beta
c =  1    #1.5 gamma
d =  0.01    #0.75 delta

beta1=0.01
beta2=0.02
beta3=0.03
beta4=0.04
beta5=0.05
beta6=0.06
beta7=0.07

"""DEFINIAMO IL SISTEMA"""
def dX_dt1(X,t=0): #X in questo caso  una lista bidimensionale
    return [  a*X[0] - beta1*X[0]*X[1]   ,  -c*X[1] + d*X[0]*X[1] ]

def dX_dt2(X,t=0): #X in questo caso  una lista bidimensionale
    return [  a*X[0] - beta2*X[0]*X[1]   ,  -c*X[1] + d*X[0]*X[1] ]

def dX_dt3(X,t=0): #X in questo caso  una lista bidimensionale
    return [  a*X[0] - beta3*X[0]*X[1]   ,  -c*X[1] + d*X[0]*X[1] ]

def dX_dt4(X,t=0): #X in questo caso  una lista bidimensionale
    return [  a*X[0] - beta4*X[0]*X[1]   ,  -c*X[1] + d*X[0]*X[1] ]

def dX_dt5(X,t=0): #X in questo caso  una lista bidimensionale
    return [  a*X[0] - beta5*X[0]*X[1]   ,  -c*X[1] + d*X[0]*X[1] ]

def dX_dt6(X,t=0): #X in questo caso  una lista bidimensionale
    return [  a*X[0] - beta6*X[0]*X[1]   ,  -c*X[1] + d*X[0]*X[1] ]

def dX_dt7(X,t=0): #X in questo caso  una lista bidimensionale
    return [  a*X[0] - beta7*X[0]*X[1]   ,  -c*X[1] + d*X[0]*X[1] ]


"""RISOLVIAMO LE EQUAZIONI DIFFERENZIALI DEL MODELLO LOTKA VOLTERRA"""
t = np.linspace(0, 20,  1000) #array temporale
X0 = [100, 50]    #condizioni iniziali delle due specie (ad ogni t, corrisponde una coppia di valori, memorizzata in nell'array X

X1 = integrate.odeint(dX_dt1,X0,t)#qui integriamo le equazioni, che abbiamo definito sopra
X2 = integrate.odeint(dX_dt2,X0,t)
X3 = integrate.odeint(dX_dt3,X0,t)
X4 = integrate.odeint(dX_dt4,X0,t)
X5 = integrate.odeint(dX_dt5,X0,t)
X6 = integrate.odeint(dX_dt6,X0,t)
X7 = integrate.odeint(dX_dt7,X0,t)

print(np.mean(X1))


prede1, predatori1 = X1.T
prede2, predatori2 = X2.T
prede3, predatori3 = X3.T
prede4, predatori4 = X4.T
prede5, predatori5 = X5.T
prede6, predatori6 = X6.T
prede7, predatori7 = X7.T

plt.plot(t, prede1 , 'b-', label='PREDE beta=0.01')
plt.plot(t, prede2 , 'r-', label='PREDE beta=0.02')
plt.plot(t, prede3 , 'g-', label='PREDE beta=0.03')
plt.plot(t, prede4, 'c-', label='PREDE beta=0.04')
plt.plot(t, prede5 , 'm-', label='PREDE beta=0.05')
plt.plot(t, prede6 , 'k-', label='PREDE beta=0.06')
plt.plot(t, prede7 , 'y-', label='PREDEbeta=0.06')

plt.grid()
plt.legend(loc='best')
plt.xlabel('tempo')
plt.ylabel('PREDE')
plt.title('ANDAMENTO PREDE')
plt.show()
```
